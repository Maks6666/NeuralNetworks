```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("churn.csv")
data
```

```python
data = data.drop("RowNumber", axis=1)
data = data.drop("CustomerId", axis=1)
data = data.drop("Surname", axis=1)
data
```

```python
# columns data types output
data.dtypes 
```

```python
# displaying unique values â€‹â€‹in columns with an object data type
for column in data:
    if data[column].dtype == "object":
        print(f"{column}: {data[column].unique()}")
```

```python
# replacing "Male" and "Female" with 1 and 0

data["Gender"].replace({"Female": 0, "Male": 1}, inplace=True)
data
```

```python
# encoding for data in "Geography" column
data = pd.get_dummies(data=data, columns=["Geography"])
data
```

```python
# visualization of the correlation between the client's account balance and the indicator of whether the client leaves the bank or not
balance_of_exited = data[data.Exited==0].Balance
balance_of_non_exited = data[data.Exited==1].Balance


plt.xlabel("Balance")
plt.ylabel("Number of Customers")
plt.title("Customers leaves visualization")



plt.hist([balance_of_non_exited, balance_of_exited], color=["red", "green"], label=["Exited", "Leave"])


# in order for label objects to be displayed correctly, you need to write a line
plt.legend()
```

```python
# visualization of the correlation between the period of ownership of a bank account (tenure) and the indicator of whether the client leaves the bank or not
tenure_of_exited = data[data.Exited==0].Tenure
tenure_of_non_exited = data[data.Exited==1].Tenure


plt.xlabel("Tenure")
plt.ylabel("Number of Customers")
plt.title("Customers tenure/leaves corelation")





plt.hist([tenure_of_non_exited, tenure_of_exited], color=["red", "green"], label=["Exited", "Leave"])

# in order for label objects to be displayed correctly, you need to write a line
plt.legend()
```

```python
# All types of data are now suitable for use by the AI â€‹â€‹model

data.dtypes 
```

```python
# using this code we convert all the data from the columns "CreditScore", "Age", "Tenure", "Balance", "NumOfProducts", "EstimatedSalary"
# to rank from 0 to 1 using MinMaxScaler()

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

# create an array of columns that need to be scaled
columns_to_scale = ["CreditScore", "Age", "Tenure", "Balance", "NumOfProducts", "EstimatedSalary"]

# use the scaler object
data[columns_to_scale] = scaler.fit_transform(data[columns_to_scale])
```

```python
data

# this completes cleaning the dataset - all values â€‹â€‹are placed in a rank between 0 and 1
```

```python
x = data.drop("Exited", axis=1)
y = data["Exited"]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=5)
```

```python
print(x_train.shape)
print(y_train.shape)


print(x_test.shape)
print(y_test.shape)
```

```python
# convert test and training data into float32 format for easy feeding to the neural network

x_train = np.asarray(x_train).astype(np.float32)
y_train = np.asarray(y_train).astype(np.float32)


x_test = np.asarray(x_test).astype(np.float32)
y_test = np.asarray(y_test).astype(np.float32)
```

```python
print(x_train.shape)
print(y_train.shape)


print(x_test.shape)
print(y_test.shape)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Dense, Flatten 
from keras.models import Sequential
```

```python
model = Sequential([
    Dense(8, input_shape = (12, ), activation = "relu"),
    Dense(4, activation = "relu"),
    Dense(1, activation = "sigmoid")

])

model.compile(
    optimizer="adam",
    loss="binary_crossentropy", 
    metrics=(["accuracy"])
            )


model.fit(x_train, y_train, epochs=50)
```

```python
model.summary()
```

```python
model.evaluate(x_test, y_test)
```

```python
pred = model.predict(x_test)
pred
```

```python
y_pred = []
for i in pred:
    if i >= 0.5:
        y_pred.append(1)
    elif i < 0.5:
        y_pred.append(0)
```

```python
for i in range(10):
    print(f"{i + 1}: Actual result: {y_test[i]} \n  Predicted result: {y_pred[i]} \n")
```

```python
pred = model.predict(x_test)


data_pred = pd.DataFrame({"Actuals": y_test.squeeze(), "Predictions": y_test.squeeze()})
print(data_pred)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print(classification_report(y_test, y_pred))
```

```python
import seaborn as sn
cm = tf.math.confusion_matrix(labels=y_test, predictions=y_pred)

plt.figure(figsize=(10, 7))
sn.heatmap(cm, annot=True, fmt="d")
plt.xlabel("Predictions")
plt.ylabel("Actuals")
```
